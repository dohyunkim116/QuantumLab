```python
from qiskit import QuantumCircuit,transpile, QuantumRegister, ClassicalRegister
```

```python
from qiskit_aer import AerSimulator
```

```python
qc=QuantumCircuit(2)
qc.x(0)
qc.x(1)
qc.h(0)
#qc.barrier([0,1])
qc.cx(0,1)
```

```python
qc.draw("mpl")
```

```python
q1=QuantumRegister(1,"q1")
q2=QuantumRegister(1,"q2")
qphase=QuantumRegister(1,"qphase")
```

```python
cr=ClassicalRegister(2)
```

```python
qc=QuantumCircuit(q1,q2,qphase,cr)
```

```python
qc.draw("mpl")
```

```python
HBread=QuantumCircuit(2,name="Bread")
HBread.h(0)
HBread.h(1)
HBread.draw("mpl")
```

```python
qc.append(HBread,[0,1])
qc.draw("mpl")
```

```python
qc.x(2)
```

```python
uf=QuantumCircuit(3,name="jens")
uf.cz(1,2)
uf.cz(0,2)
```

```python
qc.append(uf,[0,1,2])
```

```python
qc.draw("mpl")
```

```python
qc.append(HBread,[0,1])
```

```python
qc.draw("mpl")
```

```python
qc.measure([0,1],[0,1])
```

```python
qc.draw("mpl")
```

```python
backend=AerSimulator()
```

```python
transpiled_circuit = transpile(qc, backend)
transpiled_circuit.barrier([0,1])
transpiled_circuit.draw('mpl')
```

```python
from qiskit.visualization import plot_histogram
# Run the transpiled circuit using the simulated backend
job = backend.run(transpiled_circuit,shots=1000)
counts = job.result().get_counts()
plot_histogram(counts)
```
